## Plotly Dash - Cara Membuat Button di Dash Plotly

Berikut ini cara membuat button di dash plotly.

```python
#memasukan modul yang digunakan
import dash
import dash_core_components as dcc
import dash_html_components as html
```

```python
#mendefinisikan app
app = dash.Dash()
```

```python
#mengisi konten dashboard
app.layout = html.Div([
    dcc.Textarea(
        value = 'Masukan komentar disini',
        style = {'width' : '50%'}
    ),
    
    html.Br(),
    html.Button('submit', id = 'submit-form')
])
```

```python
#melihat hasil dashboard
if __name__ == '__main__':
    app.run_server(port=4050)
```
