## Plotly Dash - Cara Menggunakan Markdown dengan Dash Plotly

Berikut ini cara menggunakan markdown dengan dash plotly.

```python
#memasukan modul yang digunakan
import dash
import dash_core_components as dcc
import dash_html_components as html

from datetime import datetime as dt
```

```python
#mendefinisikan app
app = dash.Dash()
```

```python
#mengisi konten dashboard
app.layout = html.Div([
    dcc.Textarea(
        value = 'Masukan komentar disini',
        style = {'width' : '50%'}
    ),
    
    html.Br(),
    html.Button('submit', id = 'submit-form'),
    
    html.Br(),
    html.Br(),
    
    dcc.DatePickerSingle(
        id = 'dt-pick-single',
        date = dt(2015, 5, 10)
    ),
    
    html.Br(),
    html.Br(),
    
    dcc.DatePickerRange(
        id = 'dt-pick-range',
        start_date = dt(2015,5,10),
        end_date_placeholder_text = 'Pilih Tanggal Terakhir'
    ),
    
    html.Br(),
    html.Br(),
    
    dcc.Markdown(
        '''
        #### Dash dan Markdown
        
        Dash Supports [Markdown](http://commonmark.org/help),
        
        Markdown is a simple way to write and format text, 
        it includes a syntax for things like **bold text** and _italics_,
        [links](http://commonmark.org/help), inline `code` snippets, lists,
        quotes, and more.
        
        '''
    )
])
```

```python
#melihat hasil dashboard
if __name__ == '__main__':
    app.run_server(port=4050)
```
