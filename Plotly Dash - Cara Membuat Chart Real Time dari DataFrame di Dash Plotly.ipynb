## Plotly Dash - Cara Membuat Chart Real Time dari DataFrame di Dash Plotly

Berikut ini cara membuat chart real time dari dataFrame di Dash Plotly.

```python
#memasukan modul yang digunakan
import pandas as pd
import numpy as np
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objs as go
```

```python
#mendefinisikan app
app = dash.Dash()
```

```python
#membuat dataFrame
data = { 'sales' : pd.Series(np.random.randint(1,101,100)), 
'profit' : pd.Series(np.random.randint(1,1000,100)) 
}

df = pd.DataFrame(data)
df.head()
```

```python
#mengisi konten pada dashboard
app.layout = html.Div([
    
    dcc.Graph(
        id = 'Scatterplot_Chart',
        figure = {
            'data' : [
                go.Scatter(
                    x = df['sales'],
                    y = df['profit'], 
                    mode = 'markers'
                )
            ], 
            'layout' : go.Layout(
                title = 'Scatterplot of Random 60 points',
                xaxis = {'title' : 'Sales'},
                yaxis = {'title' : 'Profit'},
                hovermode = 'closest'
            )
        }
    )
])
```

```python
#melihat hasil dashboard
if __name__ == '__main__':
    app.run_server(port=4051)
```
