## Plotly Dash - Cara Membuat Scatter Plot di Dash Plotly

Berikut ini cara membuat scatter plot di dash plotly.

```python
#memasukan modul yang digunakan
import dash
import dash_core_components as dcc
import dash_html_components as html
import numpy as np
import plotly.graph_objs as go
```

```python
#mendefinisikan app
app = dash.Dash()
```

```python
#membuat data random
np.random.seed(50)
x_rand = np.random.randint(1,61,60)
y_rand = np.random.randint(1,61,60)
```

```python
#mengisi konten pada dashboard
app.layout = html.Div([
    
    dcc.Graph(
        id = 'Scatter_Chart',
        figure = {
            'data' : [
                go.Scatter(
                x = x_rand,
                y = y_rand,
                mode = 'markers'
                )
            ],
            'layout' : go.Layout(
                title = 'Scatterplot of Random 60 Points',
                xaxis = {'title' : 'Random X values'},
                yaxis = {'title' : 'Random Y values'}
            )
        }
    )
])
```

```python
#melihat hasil dashboard
if __name__ == '__main__':
    app.run_server(port = 4051)
```
