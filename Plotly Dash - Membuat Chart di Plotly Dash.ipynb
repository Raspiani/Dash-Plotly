## Plotly Dash - Membuat Chart di Plotly Dash

Berikut ini cara Membuat Chart di Plotly Dash.

```python
#memasukan modul yang digunakan
import dash
import dash_core_components as dcc
import dash_html_components as html
```

```python
#mendefinisikan app
app = dash.Dash()
```

```python
#mengisi konten pada dashboard
app.layout = html.Div([
    html.H1("Hello Raspi!!!"),
    html.Div("Selamat Belajar Dash ya!!"),
    html.Div("Dash - A Data product development framework from plotly"), 

    dcc.Graph(
        id = "simple-chart",
        figure = {
            'data' : [
                #bar chart pertama
                { 'x' : [4,6,8], 'y' : [12,16,18], 'type' : 'bar', 'name' : 'first chart' }, 
                #bar chart kedua
                { 'x' : [4,6,8], 'y' : [20,24,26], 'type' : 'bar', 'name' : 'second chart' }
            ],
            'layout' : {
                'title' : 'Simple Bar Chart'
            }
        }        
        )
])
```

```python
#melihat hasil dashboard
if __name__ == '__main__':
    app.run_server(port = 4050)
```
