## Plotly Dash - Cara Membuat Sebuah Dropdown Selector Component di Dash Plotly

Berikut ini cara membuat sebuah dropdown selector component di Dash Plotly.

```python
#memasukan modul yang digunakan
import dash
import dash_core_components as dcc
import dash_html_components as html
```

```python
#mendefinisikan app
app = dash.Dash()
```

```python
#mengisi konten pada dashboard
app.layout = html.Div([
    
    dcc.Dropdown(
        id = 'first-dropdown',
        options = [
            {'label' : 'San Francisco', 'value' : 'SF'},
            {'label' : 'New York City', 'value' : 'NYC'},
            {'label' : 'Raleigh Durham', 'value' : 'RDU'}
        ], 
        value = 'NYC'
    )
])
```

```python
#melihat hasil dashboard
if __name__ == '__main__':
    app.run_server(port=4050)
```
