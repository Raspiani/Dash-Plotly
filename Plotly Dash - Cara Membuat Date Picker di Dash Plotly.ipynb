## Plotly Dash - Cara Membuat Date Picker di Dash Plotly

Berikut ini cara membuat date picker di dash plotly.

```python
#memasukan modul yang digunakan
import dash
import dash_core_components as dcc
import dash_html_components as html

from datetime import datetime as dt
```

```python
#mendefinisikan app
app = dash.Dash()
```

```python
#mengisi konten dashboard
app.layout = html.Div([
    dcc.Textarea(
        value = 'Masukan komentar disini',
        style = {'width' : '50%'}
    ),
    
    html.Br(),
    html.Button('submit', id = 'submit-form'),
    
    html.Br(),
    html.Br(),
    
    dcc.DatePickerSingle(
        id = 'dt-pick-single',
        date = dt(2015, 5, 10)
    )
])
```

```python
#melihat hasil dashboard
if __name__ == '__main__':
    app.run_server(port=4050)
```
