## Plotly Dash - Pengenalan Plotly Dash Web Application Development Framework

Berikut ini pengenalan tentang plotly dash web application development framework.

```python
#memasukan modul yang digunakan
import dash
import dash_core_components as dcc
import dash_html_components as html
```

```python
#mendefinisikan app
app = dash.Dash()
```

```python
#mengisi konten pada dashboard
app.layout = html.Div([
    html.H1("Hello Raspi!!!"),
    html.Div("Selamat Belajar Dash ya!!"),
    html.Div("Dash - A Data product development framework from plotly")
])
```

```python
#melihat hasil dashboard
if __name__ == '__main__':
    app.run_server(port = 4050)
```
