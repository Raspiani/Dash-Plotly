## Plotly Dash - Cara Membuat Format Layout pada Chart di Dash Plotly

Berikut ini cara Membuat format layout pada chart di dash plotly.

```python
#memasukan modul yang digunakan
import dash
import dash_core_components as dcc
import dash_html_components as html
```

```python
#mendefinisikan app
app = dash.Dash()
```

```python
#mengisi konten pada dashboard
app.layout = html.Div([
    html.H1(children = "Hello Raspi!!!", 
            style = {
                'textAlign' : 'center',
                'color' : '#FF1493'
            }
           ),
    html.Div(children = "Selamat Belajar Dash ya!!", 
             style = {
                'textAlign' : 'center',
                'color' : '#FF1493'
            } 
            ),
    html.Div(children = "Dash - A Data product development framework from plotly", 
             style = {
                'textAlign' : 'center',
                'color' : '#FF1493'
            } 
            ), 

    dcc.Graph(
        id = "simple-chart",
        figure = {
            'data' : [
                #bar chart pertama
                { 'x' : [4,6,8], 'y' : [12,16,18], 'type' : 'bar', 'name' : 'first chart' }, 
                #bar chart kedua
                { 'x' : [4,6,8], 'y' : [20,24,26], 'type' : 'bar', 'name' : 'second chart' }
            ],
            'layout' : {
                'plot_bgcolor' : '#FFE4E1', 
                'paper_bgcolor' : '#ABE4E4',
                'font' : {
                    'color' : '#FFE4E2'
                }, 
                'title' : 'Simple Bar Chart'
                
            }
        }        
        )
])
```

```python
#melihat hasil dashboard
if __name__ == '__main__':
    app.run_server(port = 4050)
```
