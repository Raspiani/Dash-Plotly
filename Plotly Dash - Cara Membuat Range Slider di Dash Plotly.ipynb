## Plotly Dash - Cara Membuat Range Slider di Dash Plotly

Berikut ini cara membuat range slider di dash plotly.

```python
#memasukan modul yang digunakan
import dash
import dash_core_components as dcc
import dash_html_components as html
```

```python
#mendefinisikan app
app = dash.Dash()
```

```python
#mengisi konten pada dashboard
app.layout = html.Div([
    html.Label('Pilih Kota'), 
    dcc.Dropdown(
        id = 'first-dropdown',
        options = [
            {'label' : 'San Francisco', 'value' : 'SF'},
            {'label' : 'New York City', 'value' : 'NYC'},
            {'label' : 'Raleigh Durham', 'value' : 'RDU', 'disabled' : True}
        ], 
        value = 'NYC',
        multi = True,
        placeholder = 'Pilih Kota Disini'
        #disabled = True 
    ),
    
    html.Label(''),
    html.Label('Ini adalah Slider'), 
    dcc.Slider(
        min = 1,
        max = 10,
        value = 5, 
        step = 0.5, 
        marks = {i : i for i in range(10)}   
    ),
    
    html.Label('Ini adalah sebuah range slider'),
    dcc.RangeSlider(
        min = 1, 
        max = 10,
        step = 0.5,
        value = [3,7],
        marks = {i: i for i in range(10)}
    )
])
```

```python
#melihat hasil dashboard
if __name__ == '__main__':
    app.run_server(port=4050)
```
